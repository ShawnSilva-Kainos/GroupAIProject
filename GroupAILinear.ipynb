```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import explained_variance_score, max_error, r2_score
import plotly.express as px
import seaborn as sns
```

```python
df = px.data.iris()
```

```python
X = df.iloc[:,2].values # petal_length
Y = df.iloc[:,3].values #petal_width
```

```python
X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state=0)
```

```python
model = LinearRegression().fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)

print("coef: "+ str(model.coef_))
print("intercept: " +str(model.intercept_))
print("train R squared: " + str(model.score(x_train, y_train)))
print("test R squared: " + str(model.score(x_test,y_test)))
print("max error: " + str(max_error(y_test, predictions)))


print("explained variance score:" + str(explained_variance_score(y_test, predictions)))
print("r2 score:" + str(r2_score(y_test, predictions)))


```
